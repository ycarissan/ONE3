### Recette de la Pâte à Crêpe

#### Ingrédients :
- 250 g de farine
- 500 ml de lait
- 4 œufs
- 2 cuillères à soupe de sucre
- 1 pincée de sel
- 50 g de beurre fondu

#### Instructions :

1. **Préparation des ingrédients :**
   - Dans un grand bol, tamisez la farine.
   - Ajoutez le sucre et le sel, puis mélangez bien.

2. **Ajout des liquides :**
   - Faites un puits au centre du mélange de farine.
   - Ajoutez les œufs un par un en mélangeant bien après chaque ajout.
   - Versez progressivement le lait tout en continuant de mélanger pour éviter les grumeaux.
   - Ajoutez le beurre fondu et mélangez jusqu'à obtenir une pâte homogène.

3. **Repos de la pâte :**
   - Couvrez le bol avec un film plastique ou un torchon propre.
   - Laissez reposer la pâte au réfrigérateur pendant environ 30 minutes. Cela permet à la pâte de se détendre et aux saveurs de se mélanger.

4. **Cuisson des crêpes :**
   - Faites chauffer une poêle à crêpes à feu moyen. Vous pouvez légèrement la graisser avec un peu de beurre ou d'huile.
   - Versez une louche de pâte dans la poêle et faites-la tourner pour bien répartir la pâte.
   - Cuire jusqu'à ce que la crêpe soit dorée d'un côté, puis retournez-la pour cuire l'autre côté.
   - Répétez l'opération jusqu'à épuisement de la pâte.

5. **Service :**
   - Servez les crêpes chaudes avec les garnitures de votre choix (sucre, confiture, Nutella, fruits, etc.).

Bon appétit !
