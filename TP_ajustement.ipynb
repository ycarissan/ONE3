## üìä Matplotlib (`matplotlib.pyplot`)

### `plt.plot(x, y, label=...)`
**R√¥le**  
Trace une courbe 2D.

**Exemple minimal**
```python

plt.plot(x, y, label="Ma super courbe")
plt.legend()
plt.show()
```

**Arguments utilis√©s**
- `x` : valeurs de l‚Äôaxe des abscisses
- `y` : valeurs de l‚Äôaxe des ordonn√©es
- `label` : texte de la l√©gende

---

### `plt.xlabel("texte")` / `plt.ylabel("texte")`
**R√¥le**  
Nomme les axes du graphique.

**Exemple minimal**
```python

```

---

### `plt.title("texte")`
**R√¥le**

**Exemple minimal**


**R√¥le**

### `plt.show()`

### `np.interp(y_val, x, y)`
**R√¥le**  
Interpolation lin√©aire d‚Äôune valeur.

**Exemple minimal**
```python
x_val = np.interp(y_val, x, y)
```
Renvoie dans `x_val` la valeur interpol√©e correspondant √† `y_val` pour la fonction `y=f(x)`.

Exemple simple

```
import numpy as np

b = [0, 1, 2, 3]     # valeurs connues
c = [0, 10, 20, 30] # valeurs associ√©es

np.interp(1.5, b, c)
```

R√©sultat :
```
15.0
```
1.5 est √† mi-chemin entre 1 et 2, donc la valeur interpol√©e est √† mi-chemin entre 10 et 20.
